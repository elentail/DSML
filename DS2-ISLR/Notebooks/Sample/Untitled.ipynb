```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sample_credit.csv')
```

```python
df.describe()
```

```python
def best_corr(args):
    temp= []
    for i,v in enumerate(args):
        if v<1.0:
            temp.append((i,v))
    temp = sorted(temp ,key=lambda x:x[1])
    idx, corrv = temp[-1]
    return (df.columns[idx],corrv)
        
df.corr().apply(best_corr)
```
